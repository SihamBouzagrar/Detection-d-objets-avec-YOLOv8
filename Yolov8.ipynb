```python
import cv2
#cv2 (OpenCV) : pour lire la vidéo et manipuler les images.
from google.colab.patches import cv2_imshow
from ultralytics import YOLO
import pandas as pd

# Charger le modèle YOLOv8
model = YOLO("yolov8n.pt")

# Ouvrir la vidéo
cap = cv2.VideoCapture("/content/2165-155327596.mp4")

frame_count = 0
all_detections = []

print(f"📹 Vidéo ouverte : {cap.isOpened()}")
print(f"🎞️ Total frames : {int(cap.get(cv2.CAP_PROP_FRAME_COUNT))}")
print(f"📐 Résolution : {int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))}x{int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))}")
print("-" * 50)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Détection sur la frame
    results = model(frame, verbose=False)

    # Annoter la frame
    annotated_frame = results[0].plot()

    # ✅ Extraire les détections de chaque frame
    boxes = results[0].boxes
    if boxes is not None and len(boxes) > 0:
        for box in boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            label = model.names[cls_id]
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            all_detections.append({
                "Frame": frame_count,
                "Objet": label,
                "Confiance": f"{conf:.2f}",
                "X1": x1, "Y1": y1,
                "X2": x2, "Y2": y2
            })

    # Afficher une frame sur 10 avec résumé
    if frame_count % 10 == 0:
        print(f"\n🖼️ Frame {frame_count} — {len(boxes) if boxes is not None else 0} objet(s) détecté(s)")
        if boxes is not None and len(boxes) > 0:
            for box in boxes:
                cls_id = int(box.cls[0])
                conf = float(box.conf[0])
                label = model.names[cls_id]
                print(f"   ✅ {label} ({conf:.0%} de confiance)")
        cv2_imshow(annotated_frame)

    frame_count += 1

cap.release()
cv2.destroyAllWindows()

# ✅ Résumé final
print("\n" + "=" * 50)
print(f"✅ Analyse terminée : {frame_count} frames traitées")

if all_detections:
    df = pd.DataFrame(all_detections)

    print(f"\n📊 Total détections : {len(df)}")
    print("\n🔢 Objets les plus détectés :")
    print(df["Objet"].value_counts().to_string())

    print("\n📋 Détails des 10 premières détections :")
    print(df.head(10).to_string(index=False))
else:
    print("⚠️ Aucun objet détecté dans la vidéo.")

```
